```python
# Необходимые функции

def get_gpt_response(prompt, style):
    if style == "batman": prompt = f"You are Batman. {prompt}"
    elif style == "joker": prompt = f"You are Joker. {prompt}"

    response = openai.chat.completions.create(model="gpt-3.5-turbo", messages=[
        {"role": "system", "content": "You are a helpful assistant."}, {"role": "user", "content": prompt}],
        max_tokens=150)

    return response.choices[0].message['content'].strip()

# Функция-обработчик для кнопки
def on_button_click(b):
    user_input = input_box.value
    # Определение стиля ответа в зависимости от содержания вопроса
    if "?" in user_input:
        style = "batman"  # Пример определения стиля
    else:
        style = "joker"  # Пример определения стиля

    # Получение ответа от GPT
    response = get_gpt_response(user_input, style)

    response_label.value = f"Response ({style.capitalize()}): {response}"

```

```python
# Создание виджетов для взаимодействия с пользователем
input_box = widgets.Textarea(value='', placeholder='Здесь введите вопрос...', description='Ответ:', disabled=False)

response_label = widgets.Label(value="")

button = widgets.Button(description="Отправить вопрос")
button.on_click(on_button_click)

display(input_box, button, response_label)
```
